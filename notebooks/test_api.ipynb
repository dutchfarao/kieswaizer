```python
def vote_retriever(Besluit_Id: str) -> dict:
    
    url = f"https://gegevensmagazijn.tweedekamer.nl/OData/v4/2.0/Stemming?$filter=Verwijderd eq false and Besluit_Id eq {Besluit_Id}"
    payload={}
    headers = {
      'Cookie': 'TS01b46c01=01e9e0fda3515fcfc76a675ee8b48e103e42c3ec1e92615f8bf87ef11a3e096e480c5141741f7da4646a4f0c448688bd8098727df4'
    }

    response = requests.request("GET", url, headers=headers, data=payload)
    response_json = response.json()
    stemming_df = pd.json_normalize(response_json['value'])

    vote_dict = pd.Series(stemming_df.Soort.values,index=stemming_df.ActorFractie).to_dict()
    return vote_dict

def agendapunt_retriever(Agendapunt_Id: str) -> str:
    
    url = f"https://gegevensmagazijn.tweedekamer.nl/OData/v4/2.0/Agendapunt?$filter=Verwijderd eq false and Id eq {Agendapunt_Id}"
    payload={}
    headers = {
      'Cookie': 'TS01b46c01=01e9e0fda3515fcfc76a675ee8b48e103e42c3ec1e92615f8bf87ef11a3e096e480c5141741f7da4646a4f0c448688bd8098727df4'
    }

    response = requests.request("GET", url, headers=headers, data=payload)
    response_json = response.json()
    agendapunt_df = pd.json_normalize(response_json['value'])

    onderwerp = agendapunt_df.Onderwerp[0]

    return onderwerp


    
```
